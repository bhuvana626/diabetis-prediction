```python
import numpy as np
import pandas as pd
```

# step-1:Importing the data set

```python
data=pd.read_csv("Diabities-210331-154610.csv")
```

```python
data.head()
```

# step-2:analysing the data

```python
data.info()
```

```python
data.drop(data.tail(1).index,inplace=True)
data
```

```python
x=data.iloc[:,1:-1]

x
```

```python
y=data.iloc[:,-1]
y
```

# step-3:splitting the data set

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x, y, test_size=0.15, random_state=30)
```

# step-4:Train the model 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
classifier1= LogisticRegression(solver='lbfgs',max_iter=1000)
classifier1.fit(X_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB  
classifier2 = GaussianNB()  
classifier2.fit(X_train, y_train)  
```

# step-5:predict the result

```python
y_pred1=classifier1.predict(X_test)
y_pred1
y_pred2=classifier2.predict(X_test)
y_pred2
```

# step-6:lets check the accuracy score

```python
accuracy=round(accuracy_score(y_pred1,y_test),2)*100
print(accuracy)
accuracy=round(accuracy_score(y_pred2,y_test),2)*100
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred1)  
print(cm)
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred2)  
print(cm)
```
